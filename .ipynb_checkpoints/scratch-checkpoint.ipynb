```python
import numpy as np
import pandas as pd
from dill import dump, load
from nba_api.stats.static import players, teams
from nba_api.stats.endpoints import leaguegamelog, leaguedashplayerstats, playergamelogs
```

# REG SEASON AND PLAYOFFS (ONE PLAYER)

```python
player_id = players.find_players_by_full_name('Ja Morant')[0]['id']
```

```python
df_reg = playergamelogs.PlayerGameLogs(
    player_id_nullable=player_id,
    season_nullable='2021-22',      
    season_type_nullable='Regular Season'
        ).get_data_frames()[0]
df_reg['PLAYOFF'] = 0

df_playoffs = playergamelogs.PlayerGameLogs(
    player_id_nullable=player_id,
    season_nullable='2021-22',      
    season_type_nullable='Playoffs'
        ).get_data_frames()[0]
df_playoffs['PLAYOFF'] = 1
```

```python
df_all = pd.concat([df_reg,df_playoffs])
```

```python
df_all.sort_values(['GAME_DATE'],ascending=False,inplace=True)
df_all
```

# REGULAR SEASON AND PLAYOFFS, EVERYONE

```python
def getplayoffandregular(year):
    df_reg = leaguegamelog.LeagueGameLog(
        season=year,
        season_type_all_star='Regular Season',
        player_or_team_abbreviation='P'
    ).get_data_frames()[0]
    df_reg['PLAYOFFS'] = 0

    df_playoffs = leaguegamelog.LeagueGameLog(
        season=year,
        season_type_all_star='Playoffs',
        player_or_team_abbreviation='P'
    ).get_data_frames()[0]
    df_playoffs['PLAYOFFS'] = 1

    df_all = pd.concat([df_reg,df_playoffs])
    df_all.sort_values(['PLAYER_NAME','GAME_DATE'], ascending=[True, False], inplace=True)
```

```python
df_all = getplayoffandregular('2021-22')
```
